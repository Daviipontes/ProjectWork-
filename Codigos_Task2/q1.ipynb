```python
import control as ct 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
num = 0.07848
den = [1,0.0402,0.000385]

# Com base no sistema, definimos a amplitude da entrada de degrau como 1m¬≥/s
A = 1.0

# Criando o sistema de controle com a fun√ß√£o de transfer√™ncia
sys = ct.tf(num, den)

# Calculando a resposta ao degrau do sistema
# Multiplicando a resposta pelo fator de amplitude A
t, sys = ct.step_response(sys*A, T=np.linspace(0, 600))

# Plotando a resposta ao degrau
plt.plot(t, sys)
plt.title('Resposta ao Degrau do Sistema')
plt.xlabel('Tempo (s)')
plt.ylabel('h2 (m)')
```
